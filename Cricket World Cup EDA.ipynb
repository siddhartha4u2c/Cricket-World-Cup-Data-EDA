```python
# Performed as a part of curriculum at Great Learning

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
players=pd.read_csv('WC_players.csv')
```

```python
odi_total=pd.read_csv('ODI_Match_totals.csv')
```

```python
players.head()
```

```python
Indian_players=players[players['Country']=="India"]
```

```python
Indian_players
```

```python
odi_total.head()
```

```python
India=odi_total[odi_total['Country']=='India']
```

```python
India
```

```python
India_win=India[India['Result']=='won']
India_win
```

```python
India_win['Opposition'].value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(India_win['Opposition'].value_counts()[0:5].keys()),list(India_win['Opposition'].value_counts()[0:5]),color='g')
plt.show()
```

```python
India_loss=India[India['Result']=='lost']
India_loss
```

```python
India_loss['Opposition'].value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(India_loss['Opposition'].value_counts()[0:5].keys()),list(India_loss['Opposition'].value_counts()[0:5]),color='red')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(India_loss['Opposition'].value_counts()),labels=list(India_loss['Opposition'].value_counts().keys()),autopct='%0.2f%%')
plt.show()
```
